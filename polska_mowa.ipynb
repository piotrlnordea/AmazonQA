### Overview of which model generates speech for which language (in total there are models for 29 different languages 🤯)

```python
import pycountry

languages = set()

for model_name in TTS.list_models():
  country_code = model_name.split("/")[1]
  country_language = pycountry.languages.get(alpha_2=country_code)
  if country_language is not None:
    print("{0}: {1}".format(country_language.name, model_name))
    languages.add(country_language.name)

print()
print("Unique Languages ({0}):".format(len(languages)))
print(languages)
```

### Generating English speech

```python
# Init TTS with the target model name
tts = TTS(model_name="tts_models/en/blizzard2013/capacitron-t2-c150_v2", progress_bar=True, gpu=True)
# Run TTS
tts.tts_to_file(text=text, file_path="intro.wav")
```
