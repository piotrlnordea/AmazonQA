## Basics

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio

preload_models()
```

```python
text_prompt2 = """
     witam, jestem Piotr . Jak się masz? Agata Duda raczej nie eksperymentuje z wyglądem.
     Przyzwyczaiła nas do krótkiej fryzury i nierzucających się w oczy stylizacji.
     Wszystko wskazuje na to, że pierwsza dama nie jest fanką radykalnych zmian zarówno w ubiorze, jak i na głowie.
     Na najnowszych zdjęciach można jednak dostrzec znaczącą zmianę w jej fryzurze. Włosy są znacznie dłuższe..
"""

text_prompt = """
     Agata Duda raczej nie eksperymentuje z wyglądem.
     Przyzwyczaiła nas do krótkiej fryzury i nierzucających się w oczy stylizacji.

"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

## Advanced examples

```python
text_prompt = """
    Buenos días Miguel. Tu colega piensa que tu alemán es extremadamente malo.
    But I suppose your english isn't terrible.
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    WOMAN: I would like an oatmilk latte please.
    MAN: Wow, that's expensive!
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    I have a silky smooth voice, and today I will tell you
    about the exercise regimen of the common sloth.
"""
audio_array = generate_audio(text_prompt, history_prompt="v2/en_speaker_1")
Audio(audio_array, rate=SAMPLE_RATE)
```
